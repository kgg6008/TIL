## 탐색적 데이터 분석(EDA)의 과정
  
탐색적 데이터 분석은 데이터의 정보를 파악하며 문제 해결에 필요한 힌트를 분석하는 과정이다.  

보통 데이터는 정제되지 않은 데이터인데 이를 Raw data라고 한다.  
실제로 이 데이터들을 보면 "답이 없다..." 라는 말 밖에....
(Not null 로 지정된 테이블에 null 값이 들어가는 경우도 확인할 수 있다.)  
  
탐색적 데이터 분석은 데이터를 열어보는 상황에서 시작되며 이때부터 아래와 같은 과정을 수행한다.
- **데이터의 출처와 주제에 대해 이해한다.**  
    - 우선 데이터가 어디에서 생성되었고, 어떻게 수집된 것인지를 이해하는 것이 먼저이다.  
      데이터의 이름, 구성 요소, 그리고 데이터의 출처와 주제 등에 대해 가장 먼저 조사하여야 한다.  
      (하지만, 실무에서는 출처 및 주제를 문서화 하지 않아서 화가 나는 경우가 대다수...)  
        
        
    
- **데이터의 크기를 알아본다.**  
    - 데이터의 대략적인 양이나 개수를 알아보는 단계이다.  
      데이터의 크기에 따라서 특별한 종류이 샘플링이 필요하기도 하고, 어떤 방식으로 데이터를 처리할지도 달라질 수 있기 때문  
        
        
  
- **데이터의 feature를 살펴본다.**  
    - feature이란 데이터를 구성하는 요소를 의미한다.  
      예를 들면 신체 검사를 기록한 데이터가 있을 때, 키와 몸무게 그리고 시력 같은 측정 요소를 feature라고 부른다.  
      feature를 살펴보는 것이 아주 중요한 과정이다.

### **데이터의 속성 탐색**
  
데이터의 외형적인 힌트를 얻었다면 이제는 데이터의 실제적인 내용을 탐색해야 한다.  
크게 feature의 속성 탐색, feature 간의 상관 관계 탐색 등이 있다.  
  
  
**feature의 속성 탐색**
  
feature의 대표적으로 탐색하는 속성은 **평균, 표쥰편차, 중앙값, 사분위 수 등 통계값**이 있다.  
feature의 속성 탐색 과정에서 가장 중요한 것은 **데이터에 질문을 던지는 것**이다.  
예를 들면, '우리 나라 남성의 평균 키는 몇이나 될까?' 와 같은 질문을 날리는 것이다.  
  
  
**feature 간의 상관 관계 탐색**  
  
feature의 상관 관계 탐색이란 여러 개의 feature가 서로에게 미치는 영향력을 알아보는 것이다.  
상관 관계에서 중요한 말이 있는데, "상관관계는 인과관계를 나타내지 않는다."라는 것이다.

### 탐색한 데이터의 시각화
  
  
마지막 단계는 탐색한 데이터의 시각화 단계이다.  
숫자로만 적혀있는 데이터의 분포 모양을 파악하는 것은 힘들다.  
데이터 시각화는 수치적 자료만 가지고 파악하기 힘든 패턴이나 인사이트를 발견하는데 유용하다.  
개인적으로 인사이트라는 단어는 별로 안좋아하여 "새로운 관점 해석"이라고 말하고 싶다.
