## Navigation and visualization for get insight

We have to correlation between features.  
There are two ways to statistically explore correlations between features.  
- First is a simple correlation analysis method that calculates the correlation coefficient when there are two features.
- Second is a multiple correlation analysis that analyzes correlations among multiple target features.

> **What is correlation analysis?**  
> Correlation analysis is linear relationship between two variables expressed by a correlation coefficient.  
> If correlation coefficient is near by 1, there are strong positive correlation, if near by -1 then it means there are negative correlation.


```python
# get correlation between two feature
# pearson is one of method to calculate correlation coefficient, it is widely used method
corr = drinks_data[['beer_servings', 'wine_servings']].corr(method='pearson')
print(corr)
```

Search targets of correlation are four feature like beer_servings, wine_servings, spirit_servings, total_liters_of_pure_alchol.  
Because those features are only different types of alcohol, but in the end, similar meaning.  
  
In above correaltion coefficient between beer_servings and wine_servings return almost 0.52.

```python
# Get correlation matrix between features
cols = ['beer_servings', 'spirit_servings', 'wine_servings', 'total_litres_of_pure_alcohol']
corr = drinks_data[cols].corr(method='pearson')
print(corr)
```

It is hard to read correlation between two features.  
So we have to show graph by seaborn or pairplot.  

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings(action='ignore')

# Abbreviate the name of columns to print graph
cols_view = ['beer', 'spirit', 'wine', 'alcohol']
sns.set(font_scale=1.5)
heatmap = sns.heatmap(corr.values,
                     cbar = True,
                     annot = True,
                     square= True,
                     fmt = '.2f',
                     annot_kws={'size':15},
                     yticklabels=cols_view,
                     xticklabels=cols_view)
plt.tight_layout()
plt.show()
```

We can find feature which name is total_litres_of_pure_alcohol have correlation between almost anoter feature.

## Analyze conceptual exploration with questions

### missing value

Before the EDA we have to preprocessing missing values which in continent feature.  
If country which don't have continent information, we define continent name of 'OT' which short for 'Others'.  


```python
# preprocessing missing values

# by pandas
drinks_data['continent'] = drinks_data['continent'].fillna('OT')
drinks_data.head(10)
```

```python
# preprocessing missing values by query

query_for_fill_na = """
select
    country,
    beer_servings,
    spirit_servings,
    wine_servings,
    total_litres_of_pure_alcohol,
    case when continent is null then 'OT' else continent end as continent
from drinks
"""
drinks_data_fill_na_by_query = pd.read_sql(query_for_fill_na, conn)
drinks_data_fill_na_by_query.head(10)
```

```python
# Visualization by Pie chart

labels = drinks_data['continent'].value_counts().index.tolist()
fracs1 = drinks_data['continent'].value_counts().values.tolist()
explode = (0, 0, 0, .25, 0, 0)
plt.pie(fracs1, explode=explode, labels=labels, autopct='%.0f%%', shadow=True)
plt.title('null data to \'OT\'')
plt.show()
```

### What is the statistical information of spirit_servings by continent?


```python
# by python
# Analysis continent by agg()
result = drinks_data.groupby('continent').spirit_servings.agg(['mean', 'min', 'max', 'sum'])
result.head()
```

agg() function do almost same function with apply, but it is a function that can set the parameter of the function in parallel to get the results of various operations for the group at the same time.  


```python
# by query
query_for_statistical_information = """
select
    case when continent is null then 'OT' else continent end as continent,
    avg(spirit_servings) as mean,
    min(spirit_servings) as min,
    max(spirit_servings) as max,
    sum(spirit_servings) as sum
from drinks
group by 1
order by 1
"""
statistical_information_by_query = pd.read_sql(query_for_statistical_information, conn, index_col='continent')
statistical_information_by_query.head()
```
